### Be careful about *equality* vs. *assignment*...

`=` and `==` have very different meanings in Python.

````python
# set x equal to 5
x = 5
````

````python
# is x equal to 5?
x == 5
````

````python
x = "some other value" # reset
````

````python
# valid. what does it do?
y = x == 5
````
