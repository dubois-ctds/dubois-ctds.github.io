### Optional argument: `group`

The `scatter` method also allows you to specify a group for each data point using the `group` keyword argument.

Say we wanted to investigate the relationship between an individual's **years of experience** and their **salary** with respect to their reported **gender**.

```python
for g in scatter_phd["Gender"].unique().to_list():
    sub = scatter_phd.filter(scatter_phd["Gender"] == g)
    plt.scatter(sub["Years of Experience"].to_numpy(), sub["Salary"].to_numpy(), label=g)
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.legend()
plt.show()
```

By plotting each gender category with a label, we see our scatter plot stratified into the different categories our data has for gender. This gives us a better insight into the trends of the relationship between years of experience and salary for each gender, rather than simply looking at all gender categories together.

### [Tutorial] Grouping with `.group_by` and `.agg`

Before we learn how to create line plots, it'll be useful to learn about Polars' `.group_by` and `.agg` methods. `.group_by` lets us group rows in a DataFrame by the values of one column. We then use `.agg` to aggregate those rows into a single row per unique value. If you group by one column and aggregate with a count, the resulting DataFrame has two columns: the unique values in that column, and the count for each value. 

```python
salary.group_by('Education Level').agg(pl.len().alias("count"))
```

You can also give `.group` an optional second argument: an _aggregation_ function (i.e., that takes in an array and outputs a value, like np.mean or np.max). This argument will call that function on each column in the group, if possible.

```python
salary.group_by('Education Level').agg(pl.mean("Salary").alias("Salary mean"), pl.mean("Years of Experience").alias("Years of Experience mean"), pl.mean("Age").alias("Age mean"))
```

Notice that `np.mean` cannot be computed on strings, so there is nothing in the `"Education Level mean"` and `"Job Title mean"` columns.

### Horizontal bar charts

A **horizontal bar chart** is used to visualize **categorical** variable values. Categorical variables are non-numbers, like names and qualities (Color, Gender, Name etc.). As we saw in lecture, categorical variables come in 2 different types: *ordinal* and *nominal*. Refer to Chapter 5 of the [course notes](https://data6.org/notes/05-variables/#variable-types) to see the difference between the two types.

For a horizontal bar chart we put the category column on the vertical (y) axis and the numeric column on the horizontal (x) axis. You'll see examples throughout this lab.

**Important note:** We first aggregate the data so the DataFrame has one row per category (e.g. with `.group_by` and `.agg`). We should not try to plot a very large DataFrame with one row per individual, or we would get a bar for every row. In the cell below and throughout the lab we use Polars to aggregate the data so that visualization is easier. Just focus on the visualization calls!

<div class="alert alert-warning">
In the context of our exploration of what factors impact salary, this visualization method will be very helpful in better understanding the breakdown of features that we're interested in, such as education level and gender, within the dataset.
</div>

```python
# Just run this cell
gender = salary.group_by("Gender").agg(pl.len().alias("count"))
gender
```


Since the `gender` DataFrame only has two columns (category and count), we can plot it with a horizontal bar chart. The plot below shows one bar for each gender category, corresponding to the number of rows in the DataFrame with that value. 

```python
plt.barh(gender["Gender"].to_numpy(), gender["count"].to_numpy())
plt.xlabel("count")
plt.ylabel("Gender")
plt.show()
```

### Multiple Columns

We can also use `barh` to see multiple statistics at once. Let's use the `barh` method to see the average number of both *age* and *years of experience* for each eduction level. We'll be using the following columns:
1. `"Years of Experience"`: The average years of experience of the employees with that education level
2. `"Age"`: The average age of the employees with that education level

Run the following cell to load an aggregated version of the dataset that contains the average statistics for each education level.

```python
education_averages = salary.group_by("Education Level").agg(pl.mean("Salary").alias("Salary mean"), pl.mean("Years of Experience").alias("Years of Experience mean"), pl.mean("Age").alias("Age mean"))
education_averages
```

To look at *only* years of experience and age average for each education level, we need to select the `"Education Level"`, `"Years of Experience Average"`, and `"Age Average"` columns. We will discuss this method in the next lab, so don't worry if it doesn't make sense for now.

```python
age_experience = pl.concat([education_averages.select(["Education Level", "Years of Experience mean", "Age mean"]).slice(1, 1), education_averages.select(["Education Level", "Years of Experience mean", "Age mean"]).slice(0, 1), education_averages.select(["Education Level", "Years of Experience mean", "Age mean"]).slice(2, 1), education_averages.select(["Education Level", "Years of Experience mean", "Age mean"]).slice(3, 1)])
age_experience
```

Run the following cell to show an example of how to create an **overlaid bar chart** with two statistics.

```python
x = range(len(age_experience))
w = 0.35
plt.barh([i - w/2 for i in x], age_experience["Years of Experience mean"].to_numpy(), height=w, label="Years of Experience mean")
plt.barh([i + w/2 for i in x], age_experience["Age mean"].to_numpy(), height=w, label="Age mean")
plt.yticks(x, age_experience["Education Level"].to_numpy())
plt.legend()
plt.show()
```

If we want different visualizations for each variable, we can set the optional `overlay` argument to `False`. The default value of `overlay` is `True`, so if you don't give it a value, you will get a plot with all the included variables at once.

```python
plt.barh(age_experience["Education Level"].to_numpy(), age_experience["Years of Experience mean"].to_numpy(), label="Years of Experience mean")
plt.barh(age_experience["Education Level"].to_numpy(), age_experience["Age mean"].to_numpy(), label="Age mean")
plt.xlabel("mean")
plt.legend()
plt.show()
```

This way, we can choose if we want to create one plot with all our information or a new plot for each piece of information!
